## Linear Regression Model

```python
from sklearn.linear_model import LinearRegression
fn=LinearRegression()
```

```python
dfcopy[['horsepower','price']]
```

```python
##All Rows of Columns price & horsepowr
X=dfcopy.iloc[:,21].values.reshape(-1,1)
Y=dfcopy.iloc[:,-1].values.reshape(-1,1)
```

```python
plt.xlabel('horsepower of Cars')
plt.ylabel('prices of Cars')
plt.scatter(X,Y)
plt.show()
```

```python
fn.fit(X,Y)
```

```python
##intercept on Y
fn.intercept_
```

```python
#slope
fn.coef_
```

```python
predict_y = fn.predict(X)
```

```python
X[:5]
```

```python
fn.predict(X[:5])
```

```python
## To Know if there is large error betwwen actual and predict
Y[:5]
```

```python
##  to Find MSE
import sklearn.metrics as mse
```

```python
##standard deviation of the residuals
np.sqrt(mse.mean_squared_error(Y,predict_y))
```

```python
##the mean of absolute differences between actual and predicted values.
mse.mean_absolute_error(Y, predict_y)
```

```python
##visualize
plt.scatter(X,Y)
plt.xlabel("Hourse power")
plt.ylabel("Price")
plt.plot(X, predict_y,c="red")
plt.show()
```

##  Two Sample T test of difference means


```python
from scipy.stats import ttest_ind
```

```python
 first=dfcopy[['make','price']][dfcopy['make']=='mitsubishi'] 
first.mean()
```

```python
second=dfcopy[['make','price']][dfcopy['make']=='honda'] 
second.mean()
```

```python
ftype=dfcopy[dfcopy['make']=='mitsubishi']
stype=dfcopy[dfcopy['make']=='honda']
```

```python
##Check if mean prices of mitsubishi car are equaL TO mean prices of honda car
ttest_ind(ftype.price,stype.price)
## if p-value >0.05, there is no significant difference
```
